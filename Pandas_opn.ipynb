```python
#concat and merging--------->Joining different dataframes
df1=pd.DataFrame({'Name':['Har','Swe','Gaye','Yuva'],'age':['23','21','41','52']})
df1
```

```python
df2=pd.DataFrame({'Name':['Swe','Swe','Gaye','Yuva'],'height':['5.5','5.3','5.3','5.7']})
df2
```

```python
d=pd.concat([df1,df2])  #default adds in row
d 
```

```python
e=pd.concat([df1,df2],axis=1) #default axis=0(row) axis=1(column) doesnot check any condition it will simply add but in below case you can see height of har is missing but concat just simply added it
e
```

```python
#merging
#inner---default inner
df1=pd.DataFrame({'Name':['Har','Swe','Gaye','Yuva'],'age':['23','21','41','52']})
df1
df2=pd.DataFrame({'Name':['Har','George','Kennedy','Yuva'],'age':['23','21','41','53']})
df2
f=pd.merge(df1,df2) #by default innerjoin so we get only common elements checks both age and name
f
```

```python
g=pd.merge(df1,df2,on='Name') #checks common elements only based on age
g
```

```python
h=pd.merge(df1,df2,on='age') #checks common elements only based on age
h
```

```python
i=pd.merge(df1,df2,on='Name',how='left') #left join--alll left element plus common on right
i
```

```python
j=pd.merge(df1,df2,on='Name',how='right') #right join--alll right element plus common on left
j
```

```python
k=pd.merge(df1,df2,on='Name',how='outer') #outer join--alll right element plus common on left plus all elem on left
k
```
